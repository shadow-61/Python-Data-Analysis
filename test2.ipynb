## 数据评估

对数据进行评估，包括两方面：数据的整洁度和干净度。数据的整洁度必须同时满足三个方面：每一列为一个变量属性，每一行为一个实例
每个单元格包含一个数值。而数据的干净度可以从缺失数据、重复数据、不一致数据、无效数据和错误数据这几方面评估

## 数据的干净度

**丢失数据**

```python
original_data.info()
```

Description:产品名称,CustomerID:客户编号这两列变量数量均少于行数，说明存在缺失值。InvoiceDate: 发票日期和时间和CustomerID: 客户编号
格式需要调整

先调出Description:产品名称缺失值查看

```python
original_data[original_data["Description"].isnull()]
```

当列Description为缺失值时，UnitPrice单价似乎为0，进一步验证一下猜想

```python
original_data[(original_data["Description"].isnull())&(original_data["UnitPrice"]!=0)]
```

通过验证证明当列Description为缺失值时，UnitPrice单价为0，而数据分析的目的是，根据市场销售数据，挖掘畅销产品，如果Description

然后调出CustomerID:客户编号缺失值查看

```python
original_data[original_data["CustomerID"].isnull()]
```

此次数据分析的目的是，根据市场销售数据，挖掘畅销产品，而CustomerID:客户编号为非必要数据，故不需对其缺失值进行删除

**不一致数据**

从“国家”这一列数据分析，United Kingdom、UK、U.K.表示英国，USA、United States 表示美国，存在同一目标表达不一致的情况，需要对其
统一表示

## 无效数据或错误数据

```python
original_data.describe()
```

Quantity: 产品在交易中的数量和UnitPrice: 单价应为正数，但是这个两列存在个别数值为负数情况，应对其进行进一步分析
先找出列Quantity为负数值的数据

```python
original_data[original_data["Quantity"]<0]
```

初步分析，当列Quantity为负数值时，疑似InvoiceNo发票号码均为C开头，即表示这笔交易被取消，需要进一步验证是否存在当Quantity为负数值时
，发票号码开头不为C开头

```python
original_data[(original_data["Quantity"]<0)&(original_data["InvoiceNo"].str[0]!="C")]
```

当Quantity为负数时，存在发票号码不为C开头的情况，需要再进一步验证当Quantity为负数，发票号码不为C开头时，UnitPrice是否均为0

```python
original_data[(original_data["Quantity"]<0)&(original_data["InvoiceNo"].str[0]!="C")&(original_data["UnitPrice"]!=0)]
```

验证说明当Quantity为负数，发票号码不为C开头时，UnitPrice均为0，而此数据分析的目的是，根据市场销售数据，挖掘畅销产品,UnitPrice为0
的数据对目标分析没有意义，故需要删除无效数据

找出列UnitPrice为负数值的数据

```python
original_data[original_data["UnitPrice"]<0]
```

当UnitPrice为负数时，Description表示坏账调整的意思，不属于正常交易，所以当UnitPrice为负数时，对分析目标没有意义，故应删除

## 数据清洗

**通过上述对数据的评估，应对original_data作一下清洗**

1、删除列Description缺失对应的数据  
2、United Kingdom、UK、U.K.均表示英国，用United Kingdom统一表示，USA、United States 均表示美国，用USA 统一表示  
3、删除Quantity为负数的数据  
4、删除UnitPrice为负数的数据  
5、InvoiceDate: 发票日期和时间和CustomerID: 客户编号 格式需要调整

```python
cleaned_data=original_data.copy()
cleaned_data
```

删除列Description缺失对应的数据

```python
cleaned_data=cleaned_data.dropna(subset=["Description"])
cleaned_data["Description"].isnull().sum()
```

United Kingdom、UK、U.K.均表示英国，用United Kingdom统一表示，USA、United States 均表示美国，用USA 统一表示

```python
cleaned_data=cleaned_data.replace({"UK":"United Kingdom","U.K.":"United Kingdom","United States":"USA"})
cleaned_data["Country"].value_counts()
```

删除Quantity为负数的数据

```python
cleaned_data=cleaned_data[cleaned_data["Quantity"]>0]
cleaned_data[cleaned_data["Quantity"]<0]
```

删除UnitPrice为负数的数据

```python
cleaned_data=cleaned_data[cleaned_data["UnitPrice"]>0]
cleaned_data[cleaned_data["UnitPrice"]<0]
```

InvoiceDate: 发票日期和时间和CustomerID: 客户编号 格式需要调整

```python
cleaned_data["CustomerID"]=cleaned_data["CustomerID"].astype(str)
cleaned_data["InvoiceDate"]=pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data.info()
```
